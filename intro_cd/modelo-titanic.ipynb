# Machine Learning - Titanic

Projeto do curso de introdução à ciência de dados. O objetivo é treinar modelos de machine learning usando a biblioteca sckikit-learn. O problema é de classificação onde, baseado-se nas características dos passageiros do titanic, desejamos treinar um modelo capaz de dizer se um passageiro sobreviveu ou não ao naufrágio. Os algoritmo que usaremos será a Random Forest com teste usando split de dados.


```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
```

```python
dados = pd.read_csv("~/DIO/Carrefour-DE/intro_cd/datasets/train.csv")
```

```python
dados.head()
```

```python
dados = dados.drop(['Name', 'Ticket', 'Cabin', 'Embarked'], axis=1)
```

```python
dados.head()
```

```python
dados = dados.set_index(['PassengerId'])
```

```python
dados = dados.rename(columns={'Survived':'target'}, inplace=False)
```

```python
dados.head()
```

```python
dados.describe()
```

```python
dados.describe(include=['object'])
```

```python
dados['Sex_F'] = np.where(dados['Sex'] == 'female', 1, 0)
dados['Pclass_1'] = np.where(dados['Pclass']==1, 1, 0)
dados['Pclass_2'] = np.where(dados['Pclass']==2, 1, 0)
dados['Pclass_3'] = np.where(dados['Pclass']==3, 1, 0)
```

```python
dados.head()
```

```python
dados = dados.drop(columns=['Pclass', 'Sex'], axis = 1)
```

```python
dados.head()
```

```python
dados.isnull().sum()
```

```python
dados.fillna(0, inplace=True)
```

```python
dados.isnull().sum()
```

```python
x_train, x_test, y_train, y_test = train_test_split(dados.drop(['target'], axis=1),
                                                    dados['target'],
                                                    test_size=0.3,
                                                    random_state=1234)
```

```python
[{'treino':x_train.shape, 'teste':x_train.shape}]
```

```python
rndforest = RandomForestClassifier(n_estimators=1000,
                                    criterion='gini',
                                    max_depth=5)
```

```python
rndforest.fit(x_train, y_train)
```

```python
prob = rndforest.predict_proba(dados.drop(['target'], axis=1))[:,1]
```

```python
clas = rndforest.predict(dados.drop(['target'], axis=1))
```

```python
dados['prob'] = prob 
dados['clas'] = clas
```
